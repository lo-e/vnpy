```python
from datetime import datetime, timedelta
from time import sleep

import plotly.graph_objects as go
from jqdatasdk import *
from scipy import interpolate

from vnpy.app.script_trader import init_cli_trading
from vnpy.gateway.ctp import CtpGateway
from vnpy.trader.utility import load_json
from vnpy.trader.constant import OptionType
from vnpy.app.option_master.pricing.black_76_cython import calculate_impv
from vnpy.trader.setting import SETTINGS
```

```python
# 初始化聚宽
if not is_auth():
    auth('18521705317', '970720699')
    
# 连接接口
engine = init_cli_trading([CtpGateway])
ctp_setting = load_json('connect_ctp.json')
engine.connect_gateway(ctp_setting, 'CTP')
```

```python
# 期权品种
option_start = 'IO'
# 折现利率
r = 0.02
# 构建期权数据映射
option_portfolio = {}
for contract in engine.get_all_contracts():
    # 筛选指定期权
    if not contract.symbol.startswith(option_start) or not contract.option_underlying:
        continue
    
    """
    if contract.option_underlying != 'TA101':
        continue
    """
    
    # 订阅期权行情
    engine.subscribe([contract.vt_symbol])
    # 获取期权链字典
    chain = option_portfolio.get(contract.option_underlying, None)
    if not chain:
        chain = {
            'call':{},
            'put':{}
        }
        option_portfolio[contract.option_underlying] = chain
    
    # 区分CALL/PUT
    temp = contract.symbol.replace(contract.option_underlying, '')
    if 'C' in temp:
        chain['call'][contract.option_strike] = contract
    else:
        chain['put'][contract.option_strike] = contract
print(list(option_portfolio.keys()))
```

```python
# 绘制二维图形
def showImpv():
    # 计算合成期货的价格
    synthetic_data = {}
    current_dt = datetime.now()
    for option_underlying, chain in option_portfolio.items():

        calls = chain['call']
        puts = chain['put']

        strikes = list(calls.keys())
        strikes.sort()

        min_price_diff = -1
        atm_strike = 0
        synthetic_price = 0

        for strike in strikes:
            call = calls[strike]
            put = puts[strike]

            call_tick = engine.get_tick(call.vt_symbol)
            put_tick = engine.get_tick(put.vt_symbol)

            if not (call_tick.bid_price_1 and call_tick.ask_price_1 and put_tick.bid_price_1 and put_tick.ask_price_1):
                continue

            call_mid_price = (call_tick.bid_price_1 + call_tick.ask_price_1) / 2.0
            put_mid_price = (put_tick.bid_price_1 + put_tick.ask_price_1) / 2.0
            price_diff = abs(call_mid_price - put_mid_price)

            if min_price_diff < 0 or price_diff < min_price_diff:
                min_price_diff = price_diff
                atm_strike = strike
                synthetic_price = call_mid_price - put_mid_price + strike

        synthetic_data[option_underlying] = synthetic_price
        #print(option_underlying, atm_strike, synthetic_price, '\n')

    # 计算隐含波动率
    impv_data = {}
    t_data = {}
    for option_underlying, chain in option_portfolio.items():
        calls = chain['call']
        puts = chain['put']

        strikes = list(calls.keys())
        strikes.sort()

        # 合成期货价格
        synthetic_price = synthetic_data[option_underlying]
        # 到期时间计算
        option_expiry = list(calls.values())[0].option_expiry

        # 方案一（使用聚宽API获取交易日天数，但是仅局限于年底前）
        end_date = option_expiry.strftime('%Y%m%d')
        start_date = datetime.now().strftime('%Y%m%d')
        trading_days = get_trade_days(start_date=start_date, end_date=end_date)
        if len(trading_days) <= 1:
            continue
        t = len(trading_days) / 240
        #print(option_underlying, option_expiry.strftime('%Y-%m-%d'), len(trading_days), t, '\n')

        # 方案二（计算自然日天数）
        #"""
        cal_days = (option_expiry - datetime.now()).days
        if cal_days <= 1:
            continue
        t = cal_days / 365
        #print(option_underlying, option_expiry, cal_days, t, '\n')
        #"""
        t_data[option_underlying] = t


        data = {}
        for strike in strikes:
            call = calls[strike]
            put = puts[strike]

            if strike > synthetic_price:
                option_tick = engine.get_tick(call.vt_symbol)
                cp = 1
            else:
                option_tick = engine.get_tick(put.vt_symbol)
                cp = -1

            option_mid_price = (option_tick.bid_price_1 + option_tick.ask_price_1) / 2.0
            option_impv = calculate_impv(option_mid_price, synthetic_price, strike, r, t, cp)
            #print(strike, option_mid_price, synthetic_price, r, t, cp, option_impv)
            data[strike] = option_impv
        impv_data[option_underlying] = data

    # ====== 剔除异常数据 ======
    # 因为市场数据不足导致的合成期货价格异常
    for option_underlying, synthetic_price in synthetic_data.items():
        if not synthetic_price and option_underlying in impv_data:
            impv_data.pop(option_underlying)

    #"""
    # 隐含波动率计算结果为0
    for option_underlying, data in impv_data.items():
        invalid_impv_strikes = set()
        for strike, impv in data.items():
            if impv <= 0:
                invalid_impv_strikes.add(strike)
        for strike in invalid_impv_strikes:
            data.pop(strike)
    #"""

    # 初始化绘图
    fig = go.FigureWidget()
    for option_underlying, data in impv_data.items():
        strikes = list(data.keys())
        impvs = list(data.values())
        impv_line = go.Scatter(
            x = strikes,
            y = impvs,
            mode = 'lines',
            name = option_underlying
        )
        fig.add_trace(impv_line)

    fig.show()
    
    #"""
    # 汇总所有月份行权价
    strike_set = set()
    max_impv = 0
    min_impv = 0
    for option_underlying, data in impv_data.items():
        for strike, impv in data.items():
            strike_set.add(strike)
            if not max_impv:
                max_impv = impv
                min_impv = impv
            else:
                max_impv = max(max_impv, impv)
                min_impv = min(min_impv, impv)
    all_strikes = list(strike_set)
    all_strikes.sort()
    print(all_strikes)
    
    # 生成x, y, z数据
    x = all_strikes
    y = list(t_data.values())
    z = []
    for option_underlying, data in impv_data.items():
        strikes = list(data.keys())
        impvs = list(data.values())

        cs = interpolate.CubicSpline(strikes, impvs)
        z_buf = []
        for strike in all_strikes:
            z_value = float(cs(strike))
            z_value = max(z_value, min_impv)
            z_value = min(z_value, max_impv)
            z_buf.append(z_value)
        z.append(z_buf)
        
    # 3D曲面绘制
    fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
    fig.update_layout(
        title='波动率曲面',
        autosize = False,
        width = 600, 
        height = 600,
        scene = {
            'xaxis':{'title':'行权价'},
            'yaxis':{'title':'剩余时间'},
            'zaxis':{'title':'隐含波动率'}
        },
        margin = {'l':65, 'r':50, 'b':65, 't':90}
    )
    fig.show()
    #"""
```
